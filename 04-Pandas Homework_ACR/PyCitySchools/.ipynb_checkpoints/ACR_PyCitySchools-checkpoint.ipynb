```python
import pandas as pd
import numpy as np
```

```python
school_df = pd.read_csv("Resources/schools_complete.csv")
student_df = pd.read_csv("Resources/students_complete.csv")
```

```python
df = pd.merge(school_df, student_df, how="left", on=["school_name", "school_name"])
```

```python
# District Summary

# * Create a high level snapshot (in table form) of the district's key metrics, including:
#   * Total Schools school name unique
#   * Total Students sum of size
#   * Total Budget sum of budget per school name
#   * Average Math Score math score, mean
#   * Average Reading Score reading score, mean
#   * % Passing Math (bin to passing, non passing)
#   * % Passing Reading (bin to passing, non passing)
#   * Overall Passing Rate (Average of the above two)

df.head(2)
```

```python
total_schools = df['school_name'].unique().size
df['Total_Schools'] = total_schools
total_students = df['Student ID'].unique().size
df['Total_Students'] = total_students
total_budget = df.drop_duplicates('school_name')['budget'].sum()
df['Total_Budget'] = "${0:.0f}".format(total_budget)
avg_reading = df['reading_score'].mean()
df['Avg_Reading'] = "{0:.2f}%".format(avg_reading)
avg_math = df['math_score'].mean()
df['Avg_Math'] = "{0:.2f}%".format(avg_math)

d_summary = df[['Total_Schools', 'Total_Students', 'Total_Budget', 'Avg_Math', 'Avg_Reading']]

math_total = df[(df['math_score'] >= 70)]
math_total = math_total['math_score'].count() / total_students
d_summary['Pct_Passing_Math'] = "{0:.2%}".format(math_total)

reading_total = df[(df['reading_score'] >= 70)]
reading_total = reading_total['reading_score'].count() / total_students
d_summary['Pct_Passing_Reading'] = "{0:.2%}".format(reading_total)

overall_passing = ((math_total + reading_total) / 2)
d_summary['Pct_Passing_Overall'] = "{:.2%}".format(overall_passing)

d_summary.head(1)
```

```python
# School Summary

# * Create an overview table that summarizes key metrics about each school, including:
#   * School Name
#   * School Type
#   * Total Students
#   * Total School Budget
#   * Per Student Budget
#   * Average Math Score
#   * Average Reading Score
#   * % Passing Math
#   * % Passing Reading
#   * Overall Passing Rate (Average of the above two)
```

```python
s_summary = df.groupby('school_name')['type', 'size', 'budget','reading_score', 'math_score'].mean()
s_summary['Per Student Budget'] = s_summary['budget'] / s_summary['size']

math = df[(df['math_score'] >= 70)]
pct_math = math.groupby(['school_name'])['math_score'].count() / s_summary['size']
s_summary['Pct Passing Math'] = pct_math

reading = df[(df['reading_score'] >= 70)]
pct_reading = reading.groupby(['school_name'])['reading_score'].count() / s_summary['size']
s_summary['Pct Passing Reading']  = pct_reading

pct_overall = (pct_math + pct_reading) / 2
s_summary['Pct Passing Overall'] = pct_overall

s_summary
```

```python
df['Per Student Budget'] = df['budget'] / df['size']
df.head(2)
```

```python
# Scores by School Spending
# Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
# Average Math Score
# Average Reading Score
# % Passing Math
# % Passing Reading
# Overall Passing Rate (Average of the above two)
```

```python
bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
df["Spending Ranges Per Student"] = pd.cut(df["Per Student Budget"], bins, labels=group_names)
df.head(2)
```

```python
stu_spending = df.groupby(['Spending Ranges Per Student'])['reading_score', 'math_score'].mean()
```

```python
students_bins = df.groupby(['Spending Ranges Per Student'])['size'].count()
range_reading = df[(df['reading_score'] >= 70)]
stu_pct_reading = range_reading.groupby(['Spending Ranges Per Student'])['reading_score'].count() / students_bins
stu_spending['Pct_Passing_Reading'] = stu_pct_reading
```

```python
range_math = df[(df['math_score'] >= 70)]
stu_pct_math = range_math.groupby(['Spending Ranges Per Student'])['math_score'].count() / students_bins
stu_spending['Pct_Passing_Math'] = stu_pct_math
```
