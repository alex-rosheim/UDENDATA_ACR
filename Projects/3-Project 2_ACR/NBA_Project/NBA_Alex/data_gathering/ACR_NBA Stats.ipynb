```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
player_position = []
x=1
for id in df.PLAYER_ID:
#     x = 1
    target = 'https://stats.nba.com/player/{}/'.format(id)
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    resp = requests.get(target, headers=headers)
    soup = BeautifulSoup(resp.text, "html.parser")
    position = soup.find('span', {'class': 'player-summary__player-pos'}).get_text()
    player_position.append(position)
    print(x, id)
    x = x + 1
        
        
```

```python
type(player_position)
```

```python
player_name = list(df.PLAYER_NAME)
```

```python
player_id = list(df.PLAYER_ID)
```

```python
positiondf = pd.DataFrame(
    {'PLAYER_NAME': player_name,
     'PLAYER_ID': player_id,
     'PLAYER_POSITION': player_position
    })
```

```python
positiondf.to_csv('player_position.csv', index=False)
```
