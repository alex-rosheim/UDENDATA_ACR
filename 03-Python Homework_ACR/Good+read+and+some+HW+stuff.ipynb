```python
# avg change in profit or loss between months over the period
import csv

data = []

with open('/Users/jeff/du/UDEN201811DATA3/Week3/HW/Instructions/PyBank/Resources/budget_data.csv') as f:
    reader = csv.reader(f)
    for row in reader:
        data.append(row)
        
# how can I track all my diffs
```

```python
pairs = []
for index, row in enumerate(data):
    try:
        pairs.append((row[1], data[index + 1][1]))
    except IndexError:
        pass
```

```python
diffs = []

for pair in pairs[1:]: # iterating over list of tuples
    first, second = pair
    diff = int(second) - int(first)
    diffs.append(diff)
```

```python
import numpy as np
```

```python
np.mean(diffs)
```
